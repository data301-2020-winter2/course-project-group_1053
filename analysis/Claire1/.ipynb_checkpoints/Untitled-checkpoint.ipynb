```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
import os
os.chdir('..')
from scripts.project_functions import load_and_process
from scripts.project_functions import add_medalpts
```

```python
df = load_and_process("../data/raw/Summer-Olympic-medals-1976-to-2008 2.csv")
df.head()
```

```python

add_medalpts(df)
df = df.copy().drop(['Country_Code', 'City', 'Athlete'], axis=1)
df.head()
```

```python
import pandas_profiling as pdp
df.profile_report()
```

```python
anualpts = (df.groupby(['Country', 'Year']).Medal_Points.sum()
            .sort_values(ascending=False)
            .reset_index())
anualpts
```

```python
total = df.groupby(["Country","Year"])["Medal"].count().reset_index(name="Count").sort_values(by='Count',ascending=False).reset_index()

total
```

```python
totalpts = (df.groupby('Country')
            .Medal_Points.sum()
            .sort_values(ascending=False)
            .reset_index())
totalpts
```

```python
list(totalpts.Country[:10])
```

```python
sns.lineplot(x='Year', y='Medal_Points', hue = 'Country', data=anualpts[anualpts['Country'].isin(list(totalpts.Country[:10]))])
plt.legend(loc="center right", bbox_to_anchor=(1.4, 0.5))
```

```python
genders = pd.crosstab(df['Discipline'],df['Gender']).apply(lambda x: x/x.sum(), axis=1)
genders = genders.sort_values(by=['Men'], ascending=True)
genders
```

```python

list(genders.columns)
```

```python
ax = genders.plot(figsize=(8, 12), kind='barh', stacked='True')
ax.axvline(0.5, ls=':', color='black')
plt.legend(loc="upper left")
```

```python
genders_over_time = pd.crosstab(df['Year'],df['Gender']).apply(lambda x: x/x.sum(), axis=1)
genders_over_time
```

```python
gender_percent_time = sns.lineplot(data=genders_over_time)
gender_percent_time.lines[1].set_linestyle("-")
gender_percent_time.legend().get_lines()[1].set_linestyle("-")
gender_percent_time.axhline(0.5, ls=':', color='black')
```
